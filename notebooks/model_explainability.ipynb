```python
import pandas as pd
import matplotlib.pyplot as plt

# Load detected events
df = pd.read_csv("../data/processed/detected_events.csv")

# Compute change magnitude for explainability
df["ChangeMagnitude"] = df["MeanAfter"] - df["MeanBefore"]
df_sorted = df.sort_values(by="ChangeMagnitude", ascending=False)

# Show top 5 most impactful change points
print(df_sorted.head())

# Optional: plot top 10 magnitudes
top10 = df_sorted.head(10)
plt.bar(top10["Date"], top10["ChangeMagnitude"])
plt.xticks(rotation=45)
plt.title("Top 10 Change Points by Magnitude")
plt.ylabel("Price Change ($/bbl)")
plt.xlabel("Date")
plt.tight_layout()
plt.show()
```
